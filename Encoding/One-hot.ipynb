```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
# load data set
data = pd.read_csv('carprices.csv')
```

```python
data['Car Model'].value_counts() #
```

```python
dummies = pd.get_dummies(data['Car Model'])
dummies
```

```python

merged = pd.concat([data,dummies], axis=1)
merged
```

```python
data1 = merged.drop(['Car Model'], axis=1)
data1

y=data1.pop("Sell Price")
X=data1
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.predict([[45000,8,1,0,0,0]])
```

```python

# Evaluating the Model Performance
from sklearn.metrics import r2_score
r2_score
```
